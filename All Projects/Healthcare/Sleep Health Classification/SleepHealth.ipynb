## Using Machine Learning to create prediction model

Now that we have a better understanding of our data, we can look at different machine learning tools to use on the dataset. For this example, lets use classfication supervised machine learning to our advantage and create a model that can predict if the person have a sleeping disorder or not.

Firstly, we will be importing and downloading nessceary libraries and tools.

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report

sklearn.set_config(display='diagram')# Gives Better visualization of our pipeline
```

Lets read our file once more:

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df
```

We will be classifying whether the person has a sleep disorder or not. This will be a binary classifcation. A binary classfication is an algorithm that classify input data in only two possible categories. These two categories can be simply as class 1 and class 0.

To start off, we will be adding a new numeric column in 0s and 1s that translate to 1 meaning that the person does not have a disorder and 0 meaning that the person does have a disorder. We can achieve this by using pandas **get_dummies()**.

```python
dummies = pd.get_dummies(df, columns=['Sleep Disorder'])[['Sleep Disorder_None']]
dummies.head()# let see if it works
```

Lets check if the function worked by using**head()** function to see the first few rows of the data.

```python
df['No_Disorder'] = dummies['Sleep Disorder_None'] # Only looking at the rows with no disorder
df[dummies['Sleep Disorder_None'] ==1].head()
```

Next, we can use the function **corr()** to take a look at the correlation between other columns, and whether that person has a sleep disorder. We can only use this function with numeric values.

```python
df.corr()['No_Disorder']*100

#In future version, the corr() function can be used with non-numeric column as well.
#If we want to only select numeric columns, you may use this:
#df.corr(numeric_only=True)['No_Disorder'] * 100
```

This function allows us to see the value of correlation of each columns to our No_Disorder column. Positive correlation means that the variables move in the same direction, and negative variable moves in the opposite of our indicated column. For example, as Sleep Duration and Quality of sleep shows a strong positive correlation, meaning the higher of those values, the more likely that sample has no disorder.

##Transforming our Data

The second step is to transform our data into numeric data for our model to understand.


```python
def transform_to_numeric(data):
    d1 = df.drop(columns=['Person ID', 'Sleep Disorder']).fillna(0) #We won't be needing the person ID and Sleep disorder column (as we have already haev a binary sleep disorder column), and we will fill all NA values with 0
    return pd.get_dummies(d1).replace({True: 1, False: 0})

numeric = transform_to_numeric(df)
numeric.head()
```

```python
# We have to seperate our inputs and outputs
X = numeric.drop(columns = 'No_Disorder')
y = numeric['No_Disorder']
y.value_counts() # shows the amount of people with disorder and without in the dataset
```

Note: We can see that there is 64 more people without disorder compared to with disorder. We can keep this in mind if our model turns out to be more biased  towards predicting people without sleep disorder.

##Training our model

The first step is to train our model. We will need to seperate our dataset in to a training data and testing data.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2) # 80% of our data will be training, with the remaining 20% will be for our testing data.
```

For this classification algorithm, we will be using logistic regression. **LogisticRegressionCV** will automatically find the best hyperparameters for our model by using cross validation.

Moreover, we will be using **StandardScaler** as our columns have different scales such as age and daily steps. We have to scale our data before creating our model.

We will incorporate both of these into a **Pipeline** class. Pipeline allows us to specify the preprocessing and the model in one object. This helps enlimiate errors in our preprocessing steps.



```python
clf = make_pipeline(StandardScaler(), LogisticRegressionCV(max_iter=1500)) # Scaling our data (as our columns varies in values). Using pipeline to apply the scalling in both our training and testing data
clf.fit(X_train, y_train)
```

Furthermore, logistic regression allow us to use the **score()** function to evaluate the accuracy of our model.

```python
clf.score(X_test, y_test) #this gives us the average accuracy, the percentage of the samples we got right
```

We are getting a 93% accuracy. This is a good accuracy score!

Lets take a look at what we got wrong. We can define a function that shows the predicted and the actual value of the sample.

```python
def get_incorrect_predictions(pred, y_test):
    result = pd.DataFrame({**df.loc[X_test.index, ['Age', 'Sleep Duration', 'Sleep Disorder']], 'Prediction': pred, 'Actual': y_test}) # I chose age and sleep duration to see if there could be any correlation
    return result[result['Prediction'] != result['Actual']]

pred = clf.predict(X_test)
get_incorrect_predictions(pred, y_test)
```

The results shows that our model incorrectly predicted 4 sample to have no disorder and 1 sample to have disorder.

Note: If you remember earlier, where we looked at the distribution of our data. There was less samples on people with disorder causing our model to have a harder time predicting people with disorder compared to no disorder.
