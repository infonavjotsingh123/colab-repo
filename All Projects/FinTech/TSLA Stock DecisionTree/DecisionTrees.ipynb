## Introduction of Decision Trees on Finance Data

Decision tress are very commonly used form of machine learning algorithm for classification and regression. The decision making process is supposed to mimic and be similar to how we make choices in our everyday lives. 

A decision tree is not always going to be providing the user with an answer or a clear decision. However, it represents the data in a way so that the user can make the decision on their own. 

### Structure of a Decision Tree

A decision tree is structured with the following components:

- Root Node: The stem (base) of the tree
- Leaf Node: sub-nodes does not further split into other additional sub-nodes; represents possible outcomes
- Decision Nodes: When a sub-node is further split into other Decision nodes
- Splitting: The process of dividing a decision node or root node into other addtional decision nodes
- Depth and height: the total nu,ber of edge paths from the root node to the farthest leaf node

A decision tree looks like a upside down tree. The base of the tree being the very top node called the "root node". This root nodes flows into a series of decision nodes that represent the decisions to be made. From those decision nodes they can branch out into more decision nodes and eventually arrive at a leaf node that represents the final outcome to our choices.



The very top node as you can see is the "Root node", this root splits and branches into other "sub trees" with 2 "decision nodes" splitting from the root node. Each of the decision nodes then also split until eventually we get to a "leaf node" which is a node with no more splits.

### Advanatges and Disadvantages of a Decision Tree

A decision tree visually represents a cause and effect relationships, providing a simple view of a complex processes. They are adpatable to solve both classification and regression problems. With a decision tree, you can clarify risks, objectives and benefits.

Advantages:
- numerical or categorical data and variables works
- requires less data cleaning than other data modelling techniques
- simple structure
- models problems with multiple outputs

Disadvantages:
- noise affects data
- larger datasets are not ideal
- the decision nodes are limited to binary outcome, limiting the complexity
- can become quite complex when dealing with uncertain outcomes

### Decision tree example with code:

To achieve a better understanding to Decision Trees, let's use sklearns's base datasets to create and vizualize a Decision Tree.

````python
# Imports
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn import datasets
````

````python
#default dataset set to variable
iris = datasets.load_iris()

x = iris['data']
y = iris['target']
````

To tell you a little bit about the iris dataset, this dataset is from the defualt sklearn library. It is stuctured like a dictionary and consists of 3 different classes of irises. 

We have seperated our target variable (Y) from the rest of all of the features (X). However, we have no idea what this data is yet. This is how we can look at what this data is.

````python
colNames = iris['feature_names']
target = iris['target_names']

print(colNames)
print(target)
````

What we can see is that the feature consist of the sepal length in cm, the sepal width in cm, the petal length in cm, and petal with in cm. Our target is categorical. The Decision Tree classsifier will use the features to distinguish vetween the 3 types of 'iris' that we have. Lets create our Decision Tree classifer model:

````python
#Create object for the model
clf  = DecisionTreeClassifier(random_state=42)
# Fit model
clf.fit(x,y)
````

Lastly, we have not really got to see what this tree looks like so lets create a plot for it:

````python
#creating plot here
figure = plt.figure(figsize=(25,20))
tree = tree.plot_tree(clf, feature_names=colNames, class_names=list(target), filled=True)
````

### Predictions Using Decision Trees

What we will do in this notebook as our primary objective is to create prediction data for the next 100 days. How we will do this is to isolate a column and using a Binary Tree Classifer to generate this data. 

Firstly lets create a sub data frame that contains the column we will predict

````python
df2 = df['Open']
df2
````

Lets format this dataframe

````python
df2 = pd.DataFrame(df2)
df2.tail()
````

Next we will set up a variable to indicate how many days we are going to be prediciting into the future. 

A new column named 'Prediction' is being added to the DataFrame df2. The values in this new column are derived from the 'close' column of the same DataFrame, but these values are shifted up by 100 rows (negative future_days). 

````python
days_predicted = 100
df2['Prediction'] = df2['Open'].shift(-days_predicted)
````

````python
df2.tail()
````

Perfect, we will now create two variable to fit our model, we will need to create an X variable where we drop the prediction data. Also we are going to need our prediciton data which will be our Y variable

````python
x = np.array(df2.drop(columns = ['Prediction'] ))[:-days_predicted]
print(x)

y = np.array(df2['Prediction'])[:-days_predicted]
print(y)
````

### Decision Tree Regressor

From sklearns's standard package, we will be utilizing the Decision Tree regressor algorithm to predict our stock data 100 days out into the 'future'. Since our problem is not a classification problem we do not use the Decision Tree algorithm but the regressor since we are solving a regression problem.

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
````

````python
from sklearn.tree import DecisionTreeRegressor
# Implementing Decision Tree Regression Algorithms.
tr = DecisionTreeRegressor().fit(x_train, y_train)
````

In our new variable to seperate our data, we first drop the 'Prediction' column and also store the last 100 days that we exteneded out as a seperate array. This is where we will keep the prediction data once we have that data

````python
x_Future = df2.drop(columns = ['Prediction'])[:-days_predicted]
x_Future = x_Future.tail(days_predicted)
x_Future = np.array(x_Future)


print(x_Future)
````

We will now make the predictions using the ```.predict()``` function from sklearn. This function is used to make predictions based on the input features stored in the x_future NumPy array. This line applies the trained machine learning model (a decision tree classifier or regressor, depending on the context) to the features in x_future and stores the predicted values in the tree_prediction variable.

````python
tree_prediction = tr.predict(x_Future)
print(tree_prediction)
````

We are now going to create a subset of df2. This is to plot this 'prediction' as a continuation of our data that we currently have

````python
prediction = tree_prediction
valid = df2[x.shape[0]:]
valid['Predictions'] = prediction
````

lastly lets do the plot to see if we did our work properly

````python
plt.figure(figsize=(16,8))
plt.title("Model")
plt.xlabel('Days')
plt.ylabel('Open Price USD ($)')
plt.plot(df2['Open'])
plt.plot(valid[['Open', 'Predictions']])
plt.legend(["Original", "Valid", 'Predicted'])
plt.show()
````
